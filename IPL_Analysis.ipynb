# t-test 

#The first step when applying statistical controls is to check the normality and variance criteria

#To test for normality, apply the Shapiro-Wilk test by calling the scipy stats.shapiro() method

```python
from scipy import stats
stats.shapiro(new_data.win_by_runs.dropna())
```

```python
stats.shapiro(new_data.win_by_wickets.dropna())
```

#Since p > 0.05 for both data sets, they both conform to the normality criterion

#To test for variance, apply the Levene test by calling the scipy stats.levene() method

```python
stats.levene(new_data.win_by_runs.dropna(), new_data.win_by_wickets.dropna())
```

#As p > 0.05 the data sets conform to the variance criterion

#Independent of the sample sizes when the norm/var criteria hold true then we can proceed and apply the t-test 

```python
t, p = stats.ttest_ind(new_data.win_by_runs.dropna(), new_data.win_by_wickets.dropna())
t, p
```

#Since p > a (0.05) we say that the null hypothesis is suggested

# Hypothesis Testing

#### H0 - Null Hypothesis - Match result is not affected by toss  result 

#### H1 - Alternate Hypothesis -  Match result is affected by toss  result 

#### Significance value is assumed to be 5% that is 0.05¶

```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
import math
```

```python
toss_win_match_win=[]
for i in range(len(data)):
    if(data['toss_winner'][i]==data['winner'][i]):
        toss_win_match_win.append(1)
    else:
        toss_win_match_win.append(0)


sampleMean=np.array(toss_win_match_win).mean()
hypMean=0.5
standardPopulation=np.std(toss_win_match_win)
N=len(toss_win_match_win)
```

```python
z=(sampleMean-hypMean)/(standardPopulation/math.sqrt(N))
pval=stats.norm.cdf(z)

if pval<=0.05:
    print("Null Hypothesis(H0) Rejected")
else:
    print("Null Hypothesis(H0) Pausible")
    
pval
```
